### Using the uploaded audio file as source
Now, modify the `speech_recognition` code to use the uploaded `audio_filename` instead of `sr.Microphone()`.

```python
!pip install pydub
!apt-get update && apt-get install -y ffmpeg
import speech_recognition as sr
from pydub import AudioSegment
import os

# Initialize recognizer class
r = sr.Recognizer()

# Check if the uploaded file is already a WAV file
if audio_filename.lower().endswith('.wav'):
    converted_audio_filename = audio_filename
    print(f"File '{audio_filename}' is already a WAV file. Skipping conversion.")
elif audio_filename.lower().endswith('.mp3'):
    print(f"Converting {audio_filename} to WAV...")
    # Load the MP3 file
    audio = AudioSegment.from_mp3(audio_filename)
    # Export to a WAV file
    converted_audio_filename = "converted_audio.wav"
    audio.export(converted_audio_filename, format="wav")
    print(f"Conversion complete. Using {converted_audio_filename}.")
else:
    print(f"Unsupported audio format for file: {audio_filename}. Attempting to process as is.")
    converted_audio_filename = audio_filename

with sr.AudioFile(converted_audio_filename) as source:
    # listen for the data (load the audio into memory)
    audio_text = r.record(source)

    try:
        # using google speech recognition
        print("Text: "+r.recognize_google(audio_text))
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```
